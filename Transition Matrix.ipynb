#### 2. For each sentence, create pairs of consecutive words (suggestion: change the '.' to the word 'PERIOD')
For example："I lived there two years and two months. At present I am a sojourner in civilized life again."

**Hint1**: Given two lists, you can use the `zip` function to combine them. **Hint2**: You can create a list which includes all elements, except the first one, by using slicing: mylist[1:]

```python
# Change the "." to the word "PERIOD"
data_file_location = "input.txt"
inputfile = open(data_file_location, 'r', encoding='utf8')

lines = inputfile.readlines()
for i in range(len(lines)):
    lines[i] = lines[i].replace("."," PERIOD")

# Print the first 5 lines to check
print(lines[0:5])
```

```python
# Create a list which includes all elements
words = []
for i in range(len(lines)):
        words += lines[i].split()

# Create a list which includes all elements except the first one, combine the two lists using `zip`
words2 = words[1:]
myzip = zip(words,words2)

# Delete the tuples which start from "PERIOD" from the zipped list
mylist=[]
for i in myzip:
    if i[0]!='PERIOD':
        mylist.append(tuple(i))
        
print(mylist)
```

#### 3. Count how often each pair of words occurs in the input text and record it for later retrieval.
 A. Use a dictionary of dictionaries  B. Use a numpy matrix

```python
# Use numpy matrix: 
# Count the unique words in the inputfile
wordcount = len(set(words))
print(wordcount)
```

```python
# Generate a square matrix, equal in size to the number of unique words in the input file
import numpy as np

M = np.zeros(shape=(wordcount,wordcount))
uwords = list(set(words))
uwords.sort()

for element in mylist:
    i = uwords.index(element[0])
    j = uwords.index(element[1])
    M[i][j] += 1
```

```python
# Display the first 5 lines of the frequency matrix
print(M[0:5])
```

```python
# Count the frequency sum to check whether they are all zeros
sum = 0
for i in range(wordcount):
    for j in range(wordcount):
        sum += M[i][j]
print(sum)
```

#### 4. Normalize the counts and turn them into probabilities.
For each word, find all paired words, add up their frequencies and divide each frequency by total. For example, if "i" is associated with "lived" and "am", and they both of occur once, their total occurrence is 2 and the resulting probabilities are "lived": 0.5, "am":0.5

```python
# Convert to probabilities
for row in M:
    n = np.sum(row)
    if n > 0:
        row[:] = [f/n for f in row]
```

#### 5. Generate text based on this transition matrix
A. Pick a word at random (start with "the" if you like)
B. For the picked word, find all words it pairs with and get their probabilities. Pick the next word based on weighted probabilities. For example, if "the" is paired with "one" with probability .1 and "only" is paired with probability .3, "only" should be picked three times more likely than "one." Simplest way of doing this is via `random.choices()`.
C. Having picked a word, use logic from B to select the next word. Do this in a loop.
D. If "PERIOD" is picked, start a new sentence.
E. Generate one hundred sentences.

```python
import random

sentences = []
for i in range(100):
    word1 = random.choices(uwords)[0] # Random starting word
    sentence = word1
    while word1 != 'PERIOD': # If "RERIOD" is picked, start a new sentence
        index = uwords.index(word1)
        prob_list = M[index]
        word2 = random.choices(uwords,weights=list(prob_list))[0]
        sentence += " " + word2
        word1 = word2
    sentences.append(sentence)
```

```python
# Replace "PERIOD" by "." to make a sentence
for i in range(len(sentences)):
    sentences[i] = sentences[i].replace(" PERIOD",".")

# Generate 100 sentences
outputtext = sentences[0:100]
print(outputtext)
```

```python
file = open('output.txt','w')
for sentence in outputtext:
    file.write(sentence + '\n')
file.close()
```
